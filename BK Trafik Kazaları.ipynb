```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("2005.csv",low_memory=False)
df2=pd.read_csv("2009.csv",low_memory=False)
df3=pd.read_csv("2012.csv",low_memory=False)
```

```python
genel=pd.concat([df1,df2,df3])
```

```python
genel = genel.drop(['Accident_Index', 'Location_Easting_OSGR', 'Location_Northing_OSGR','Longitude',
       'Latitude', 'Police_Force','Date','Local_Authority_(District)','Local_Authority_(Highway)',
       '1st_Road_Class', '1st_Road_Number','Junction_Detail', 'Junction_Control', '2nd_Road_Class',
       '2nd_Road_Number', 'Pedestrian_Crossing-Human_Control', 'Pedestrian_Crossing-Physical_Facilities', 
       'Special_Conditions_at_Site','Did_Police_Officer_Attend_Scene_of_Accident',
       'LSOA_of_Accident_Location','Urban_or_Rural_Area'], axis=1)
```

```python
genel.head(10)
```

```python
genel=genel.dropna()
```

```python
yÄ±llar=genel["Year"].unique()
yÄ±l=genel["Year"].value_counts().values
fig = plt.figure(figsize =(10, 7))
plt.bar(yÄ±llar,yÄ±l)
plt.xlabel("YÄ±llar")
plt.ylabel("Kaza SayÄ±larÄ±")
plt.title("YÄ±llara gÃ¶re kaza sayÄ±larÄ±")
plt.xticks(yÄ±llar)
```

## AkÅŸam Vakti GerÃ§ekleÅŸen KazalarÄ±n Toplam Kazalara OranÄ±


```python
yÄ±llar=genel["Year"].unique()
sayÄ±_g=np.array([])
sayÄ±_a=np.array([])

for a in yÄ±llar:
    gÃ¼ndÃ¼z=genel[(genel["Year"] == a) & (genel["Time"] < "20")].values
    sayÄ±_g=np.append(sayÄ±_g,gÃ¼ndÃ¼z.shape[0])
    
for b in yÄ±llar:
    akÅŸam=genel[(genel["Year"] == b) & (genel["Time"] > "20")].values
    sayÄ±_a=np.append(sayÄ±_a,akÅŸam.shape[0])
    
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

plt.bar(yÄ±llar, sayÄ±_a, color ='r', width = barWidth)
plt.bar(yÄ±llar-0.2, sayÄ±_g, color ='g', width = barWidth)
plt.xticks(yÄ±llar)
plt.show()
```

# ZararlÄ± Madde TaÅŸÄ±yan AraÃ§larÄ±n Dahil OlduÄŸu Kaza SayÄ±larÄ±

```python
yÄ±llar=genel["Year"].unique()
kimyasal=np.array([])
for a in yÄ±llar:
    gÃ¼ndÃ¼z=genel[(genel["Year"] == a) & (genel["Carriageway_Hazards"] !="None")].values
    kimyasal=np.append(kimyasal,gÃ¼ndÃ¼z.shape[0])

fig = plt.subplots(figsize =(12, 8))
plt.bar(yÄ±llar,kimyasal)
plt.xlabel("YÄ±llar")
plt.ylabel("Kaza SayÄ±larÄ±")
plt.title("ZararlÄ± Madde TaÅŸÄ±yan AraÃ§larÄ±n Dahil OlduÄŸu Kaza SayÄ±sÄ±")


plt.xticks(yÄ±llar)
```

# Hava koÅŸullarÄ± daÄŸÄ±lÄ±mÄ±


```python
import seaborn as sns
road_cond = genel["Road_Surface_Conditions"].value_counts()

road_cond_arr = genel["Road_Surface_Conditions"].unique()
road_num_acc_arr = road_cond.values

weather_cond = genel["Weather_Conditions"].value_counts() 

weather_cond_arr = genel["Weather_Conditions"].unique()
weather_num_acc_arr = weather_cond.values
plt.figure(figsize = (20,9),facecolor='grey')
plt.subplot(1, 2, 1)

plt.pie(road_num_acc_arr, labels = road_cond_arr, colors = sns.color_palette(),startangle = 30,textprops={'size': 'large'},explode=(0.02,0.02,0.02,0.02,0.3),autopct="%1.1f%%")
plt.legend()
plt.title("Yol Durumuna GÃ¶re Kaza DaÄŸÄ±lÄ±mÄ±",weight="bold")


plt.subplot(1, 2, 2)

plt.pie(weather_num_acc_arr, labels = weather_cond_arr,colors = sns.color_palette(),startangle = 30,textprops={'size': 'large'},explode=(0.01,0.01,0.01,0.01,0.01,0.20,0.3,0.50,0.7),autopct="%1.1f%%")
plt.legend(loc ="lower left")
plt.title("Hava KoÅŸullarÄ±na GÃ¶re Kaza DaÄŸÄ±lÄ±mÄ±",weight="bold")

my_circle = plt.Circle((0,0),0.5, color ='grey')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.show()
```
