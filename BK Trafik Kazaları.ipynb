```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("2005.csv",low_memory=False)
df2=pd.read_csv("2009.csv",low_memory=False)
df3=pd.read_csv("2012.csv",low_memory=False)
```

```python
genel=pd.concat([df1,df2,df3])
```

```python
genel.head(5)
```

```python
genel.columns
```

```python
genel = genel.drop(['Accident_Index', 'Location_Easting_OSGR', 'Location_Northing_OSGR','Longitude',
       'Latitude', 'Police_Force','Date','Local_Authority_(District)','Local_Authority_(Highway)',
       '1st_Road_Class', '1st_Road_Number','Junction_Detail', 'Junction_Control', '2nd_Road_Class',
       '2nd_Road_Number', 'Pedestrian_Crossing-Human_Control', 'Pedestrian_Crossing-Physical_Facilities', 
       'Special_Conditions_at_Site','Did_Police_Officer_Attend_Scene_of_Accident',
       'LSOA_of_Accident_Location','Urban_or_Rural_Area'], axis=1)
```

```python
genel.head(10)
```

```python
genel=genel.dropna()
```

# Yıllara göre kaza sayıları

```python
genel["Year"].value_counts()
```

```python
yıllar=genel["Year"].unique()
yıl=genel["Year"].value_counts().values
fig = plt.figure(figsize =(10, 7))
plt.bar(yıllar,yıl)
plt.xlabel("Yıllar")
plt.ylabel("Kaza Sayıları")
plt.title("Yıllara göre kaza sayıları")
plt.xticks(yıllar)
```

## Akşam Vakti Gerçekleşen Kazaların Toplam Kazalara Oranı


```python
yıllar=genel["Year"].unique()
sayı_g=np.array([])
sayı_a=np.array([])

for a in yıllar:
    gündüz=genel[(genel["Year"] == a) & (genel["Time"] < "20")].values
    sayı_g=np.append(sayı_g,gündüz.shape[0])
    
for b in yıllar:
    akşam=genel[(genel["Year"] == b) & (genel["Time"] > "20")].values
    sayı_a=np.append(sayı_a,akşam.shape[0])
    
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

plt.bar(yıllar, sayı_a, color ='r', width = barWidth)
plt.bar(yıllar-0.2, sayı_g, color ='g', width = barWidth)
plt.xticks(yıllar)
plt.show()
```

# Zararlı Madde Taşıyan Araçların Dahil Olduğu Kaza Sayıları

```python
yıllar=genel["Year"].unique()
kimyasal=np.array([])
for a in yıllar:
    gündüz=genel[(genel["Year"] == a) & (genel["Carriageway_Hazards"] !="None")].values
    kimyasal=np.append(kimyasal,gündüz.shape[0])

fig = plt.subplots(figsize =(12, 8))
plt.bar(yıllar,kimyasal)
plt.xlabel("Yıllar")
plt.ylabel("Kaza Sayıları")
plt.title("Zararlı Madde Taşıyan Araçların Dahil Olduğu Kaza Sayısı")


plt.xticks(yıllar)
```

# Hava koşulları dağılımı


```python
import seaborn as sns
road_cond = genel["Road_Surface_Conditions"].value_counts()

road_cond_arr = genel["Road_Surface_Conditions"].unique()
road_num_acc_arr = road_cond.values

weather_cond = genel["Weather_Conditions"].value_counts() 

weather_cond_arr = genel["Weather_Conditions"].unique()
weather_num_acc_arr = weather_cond.values
plt.figure(figsize = (20,9),facecolor='grey')
plt.subplot(1, 2, 1)

plt.pie(road_num_acc_arr, labels = road_cond_arr, colors = sns.color_palette(),startangle = 30,textprops={'size': 'large'},explode=(0.02,0.02,0.02,0.02,0.3),autopct="%1.1f%%")
plt.legend()
plt.title("Yol Durumuna Göre Kaza Dağılımı",weight="bold")


plt.subplot(1, 2, 2)

plt.pie(weather_num_acc_arr, labels = weather_cond_arr,colors = sns.color_palette(),startangle = 30,textprops={'size': 'large'},explode=(0.01,0.01,0.01,0.01,0.01,0.20,0.3,0.50,0.7),autopct="%1.1f%%")
plt.legend(loc ="lower left")
plt.title("Hava Koşullarına Göre Kaza Dağılımı",weight="bold")

my_circle = plt.Circle((0,0),0.5, color ='grey')
fig = plt.gcf()
fig.gca().add_artist(my_circle)

plt.show()
```
